## Potencial de interação

7) Defina uma função que calcule o potencial U dados r, ε e σ.

```julia
function potential(r, epsilon, sigma)
    u = zeros(0)
    for i in 1:length(r)
        append!(u, 4*epsilon*((sigma/r[i])^12 - (sigma/r[i])^6))
    end
    return u
end
```

8) Crie um vetor de distâncias, R = [r1,....,rN] e calcule a energia potencial para cada distância, dados ε e σ. Faça um gráfico.

```julia
R = collect(0.05:0.001:0.5)
sigma = 0.1
epsilon = 1
U = potential(R, epsilon, sigma)

using Plots
plot(R,U,xlabel="distancia (r)",ylabel="potencial (U)", 
    lw=2, linecolor = :red,
    ylim = (minimum(U), 0.5), size=(400,400),framestyle=:box,
    legend = false)
```

11) Escolha valores para ε e σ tais que o mínimo de energia esteja próximo de 1.0, e some, para todos os pares de partículas (sem repetições! - quantos pares são?) o potencial U calculado. Defina uma função para isto.<p>
<font color='green'> O número de pares é dado por: $\binom{10}{2} = \frac{10!}{2!8!} = 45$. Usando $\sigma = 0.9 $ e $\varepsilon = 1$<font>

```julia
sigma = 0.9
epsilon = 1

function get_distances(p) #Lê as posições e retorna um array de distâncias
    r = zeros(0) #lista de distâncias
    for i in 1:length(p)
        for j in i+1:length(p)
            dist = sqrt((p[j][1]-p[i][1])^2 + (p[j][2]-p[i][2])^2)
            append!(r, dist)
        end
    end
    return r
end
r = get_distances(p) 
U_total = sum(potential(r, epsilon, sigma))
```

12) Procure modificar sua função para acelerar sua execução. Veja com cuidado que contas são feitas. É possível evitar a repetição de contas?

```julia
using BenchmarkTools
@btime potential($r,$epsilon,$sigma)
```

```julia
function potential2(r, epsilon, sigma)
    
    u = similar(r)
    A = 4*epsilon
    sigma6=sigma^6
    sigma12=sigma6^2
    
    for i in 1:length(r)
        r6=r[i]^6
        r12=r6^2
        u[i]=A*(sigma12/r12 - sigma6/r6)
    end
    return u
end

@btime potential2($r,$epsilon,$sigma)
```

13) Sua função, em princípio, não deve alocar quase nada de memória, já que ela faz uma conta com dados de entrada e devolve um único número. Se sua função está alocando memória, procure modificá-la para evitar isso.

 <font color='green'> Substituir "zeros" e criar o vetor com "similar" resolve o problema de alocação de memória. Precisamos testar se as funções são equivalentes<font>

```julia
using Test
@test isapprox(potential(r,epsilon,sigma), potential2(r,epsilon,sigma))
```
