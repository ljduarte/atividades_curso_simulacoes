1) Defina uma função que calcule o potencial U dados r, ε e σ.

```julia
function potential(r, epsilon, sigma)
    u = similar(r)
    for i in 1:length(r)
        u[i]=4*epsilon*((sigma/r[i])^12 - (sigma/r[i])^6)
    end
    return u
end
```

2) Crie um vetor de distâncias, R = [r1,....,rN] e calcule a energia potencial
para cada distância, dados ε e σ. Faça um gráfico.

```julia
R = collect(0.1:0.1:5)
sigma = 0.1
epsilon = 1
U = potential(R, epsilon, sigma)

using Plots
plot(R,U,xlabel="distancia (r)",ylabel="potencial (U)")
```

3) Entenda o efeito da variação de ε fazendo novos gráficos com ε variável.

```julia
sigma = 0.1
epsilon = 10
U = potential(R, epsilon, sigma)

using Plots
plot(R,U,xlabel="distancia (r)",ylabel="potencial (U)")
```
